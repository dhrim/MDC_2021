```
-1 ~ +1, 0.001

y = x^2 + sin(10x)
```


````python
import numpy as np
import matplotlib.pyplot as plt


from tensorflow import keras
from tensorflow.keras.layers import Dense, Input

````

````python
x = np.arange(-1, 1, 0.001)
np.random.shuffle(x)
y = x**2 + np.sin(x*10)

plt.scatter(x, y)
plt.show()
````

````python
index = int(len(x)*0.8)
train_x, test_x = x[:index], x[index:]
train_y, test_y = y[:index], y[index:]

plt.scatter(train_x, train_y)
plt.show()
plt.scatter(test_x, test_y)
plt.show()
````

````python
model = keras.Sequential()
model.add(Input(1))
model.add(Dense(10*10, activation="tanh"))
model.add(Dense(10*10, activation="tanh"))
model.add(Dense(1))

model.compile(loss="mse", optimizer="RMSProp", metrics=["mae"])
# SGD, AdaGrad, RMSProp

hist = model.fit(train_x, train_y, epochs=1000, verbose=0, batch_size=128, validation_split=0.1)

plt.plot(hist.history['loss'], label="loss")
plt.plot(hist.history['val_loss'], label="val_loss")
plt.legend()
plt.show()

loss, mae = model.evaluate(test_x, test_y)
print("loss =", loss)
print("mae =", mae)
````

````python
y_ = model.predict(test_x)
plt.scatter(test_x, test_y)
plt.scatter(test_x, y_)
plt.show()
````
