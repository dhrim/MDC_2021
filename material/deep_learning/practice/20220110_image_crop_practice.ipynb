```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
```

# 흑백 이미지 crop

```python
(train_x, train_y), (test_x, test_y) = tensorflow.keras.datasets.mnist.load_data()
```

```python
image = train_x[0]
plt.imshow(image, cmap="binary")
plt.show()
```

```python
print(image.shape)
```

```python
cropped = image[5:20, 5:20]
plt.imshow(cropped, cmap="binary")
plt.show()
```

# 파일로 저장/읽기

```python
plt.imsave("cropped_mnist.jpg", cropped)
```

```python
loaded = plt.imread("cropped_mnist.jpg")
print(loaded.shape)
plt.imshow(loaded)
plt.show()
```

# numpy로 저장/읽기

```python
np.save("crooped.npy", cropped)
```

```python
loaded = np.load("crooped.npy")
print(loaded.shape)
plt.imshow(loaded, cmap="binary")
plt.show()
```
