```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 데이터 준비

```python
import tensorflow as tf

(train_x, train_y), (test_x, test_y) = tf.keras.datasets.mnist.load_data()
```

```python
feature = train_x[:10000].reshape(-1, 28*28)
label = train_y[:10000]
```

# PCA

```python
from sklearn.decomposition import PCA

converted = PCA(n_components=2).fit_transform(feature)
print(converted.shape)

plt.figure(figsize=(15,15))
plt.scatter(converted[:,0], converted[:,1], c=label)
plt.colorbar()

for i in range(10):
    class_center = np.mean(converted[label == i], axis=0)
    textbox={'facecolor':'white', 'boxstyle':'round','alpha':0.5}
    plt.text(class_center[0], class_center[1], str(i), fontsize=20, bbox=textbox)
    
plt.show()
```

```python
from sklearn.decomposition import PCA
import plotly.express as px

converted = PCA(n_components=3).fit_transform(feature)
print(converted.shape)

fig = px.scatter_3d(
    converted, x=0, y=1, z=2,
    color=label, labels={'class': 0}
)
fig.update_traces(marker_size=8)
fig.show()
```

# T-SNE

```python
from sklearn.manifold import TSNE

converted = TSNE(n_components=2).fit_transform(feature)
print(converted.shape)

plt.figure(figsize=(15,15))
plt.scatter(converted[:,0], converted[:,1], c=label)
plt.colorbar()

for i in range(10):
    class_center = np.mean(converted[label == i], axis=0)
    textbox={'facecolor':'white', 'boxstyle':'round','alpha':0.5}
    plt.text(class_center[0], class_center[1], str(i), fontsize=20, bbox=textbox)
    
plt.show()
```

```python
from sklearn.manifold import TSNE
import plotly.express as px

converted = TSNE(n_components=3).fit_transform(feature)
print(converted.shape)

fig = px.scatter_3d(
    converted, x=0, y=1, z=2,
    color=label, labels={'class': 0}
)
fig.update_traces(marker_size=8)
fig.show()
```
